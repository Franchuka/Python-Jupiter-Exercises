## Groupby
The groupby method allows you to group rows of data together and call aggregate functions

```python
import pandas as pd
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

 Now you can use the .groupby() method to group rows together based off of a column name. For instance let's group based off of Company. This will create a DataFrameGroupBy object:

```python
df.groupby('Company')
```

```python
by_comp = df.groupby('Company')
```

```python
by_comp.mean()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.std()
```

```python
by_comp.describe()
```

```python
pd.DataFrame(by_comp.describe().loc['GOOG']).round(decimals=2)
```

#### Split Data into Groups
Pandas object can be split into any of their objects. There are multiple ways to split an object like −

* obj.groupby('key')
* obj.groupby(['key1','key2'])
* obj.groupby(key,axis=1)

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df2 = pd.DataFrame(ipl_data)
```

```python
df2
```

```python
# View groups
teams = df2.groupby('Team')
```

```python
teams.groups
```

```python
teams_year = df2.groupby(['Team', 'Year'])
teams_year.groups
```

```python
type(teams_year.groups)
```

```python
year = df2.groupby('Year')
year.groups
```

```python
for yr, data in year:
    print(yr)
    print(data)
```

#### Select a Group
Using the get_group() method, we can select a single group

```python
year.get_group(2014)
```

```python
year.get_group(2014)['Points'].mean()
```

```python
# Check the mean of points per year
year['Points'].mean()
```

#### Applying Multiple Aggregation Functions at Once
With grouped Series, you can also pass a list or dict of functions to do aggregation with, and generate DataFrame as output

```python
import numpy as np
teams['Points'].agg([np.mean, np.sum, np.std])
```
