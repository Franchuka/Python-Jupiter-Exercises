### Legends, Labels and Titles

Now that we have covered the basics of how to create a figure canvas, lets look at how decorate a figure with titles, axis labels and legends 

### Legends
you can use the label = ##'label text'## keyword argument when plots or other objects are added to the figure, and then using the legend method without arguments to add the legend to the figure: 

```python
fig3 = plt.figure()
ax = fig3.add_axes([0,0,1,1])
ax.plot(x, x**2, label = 'x**2')
ax.plot(x, x**3, label = 'x**3')
ax.legend(loc = 0)
```

```python
# to change the location of your legend, you use the loc argument
fig3 = plt.figure()
ax = fig3.add_axes([0,0,1,1])
ax.plot(x, x**2, label = 'x**2')
ax.plot(x, x**3, label = 'x**3')
ax.legend(loc = 5)
```

## Setting Colors, Linewidth and lineType
Matplotlib gives you a lot of options for the customization of colors, linewidths, and linetypes.
    there is the basic MATLAB like syntax (which i would suggest you avoid using for more clarity sake):

```python
#Colors with MatLab syntax
fig4 = plt.figure()
ax = fig4.add_axes([0,0,1,1])
ax.plot(x,x**2, 'b')
ax.plot(x, x**3, 'g')
```

```python
fig4 = plt.figure()
ax = fig4.add_axes([0,0,1,1])
ax.hist(x, bins = 10, color = 'orange')
```

```python
# using the color arguments
fig4 = plt.figure()
ax = fig4.add_axes([0,0,1,1])
ax.plot(x,x+1, color = 'BROWN')
ax.plot(x, x+2, color = '#8B008B')
#The code is also a way of setting colors using the rgb hex code
```

```python
# to adjust the linewidth you use the 'linewidth' or lw argument. 
#to adjust the line style we use the 'linestyle' or the ls arguments.
fig6, ax1 = plt.subplots(figsize= (12, 7))

ax1.plot(x, x+1, color = 'black', linewidth = 0.25, label = 'x+1')
ax1.plot(x, x+2, color = 'black', linewidth = 0.5, label = 'x+2')
ax1.plot(x, x+3, color = 'black', linewidth = 1, label = 'x+3')
ax1.plot(x, x+4, color = 'black', linewidth = 2, label = 'x+4')
#possible linestyles '-', '--', '-.', ':', 'steps'
ax1.plot(x, x+5, color = 'orange', lw = 2, linestyle = '-', label = 'x+5')
ax1.plot(x, x+6, color = 'orange', lw = 2, linestyle = '--', label = 'x+6')
ax1.plot(x, x+7, color = 'orange', lw = 2, linestyle = '-.', label = 'x+7')
ax1.plot(x, x+8, color = 'orange', lw = 2, linestyle = ':', label = 'x+8')
#possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', '5', '6', '7', '8', '9'
a = ax1.plot(x, x+9, color = 'green', lw = 2, linestyle = '-', marker = '+', label = 'x+9')
b = ax1.plot(x, x+10, color = 'green', lw = 2, linestyle = '--', marker = 'o', label = 'x+10')
c= ax1.plot(x, x+11, color = 'green', lw = 2, linestyle = '-.', marker = '*', label = 'x+11')
d= ax1.plot(x, x+12, color = 'green', lw = 2, linestyle = ':', marker = 's', label = 'x+12')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend(['x+'+str(x) for x in range(1,13)], loc = 0)
```
