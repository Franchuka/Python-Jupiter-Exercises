## Groupby() Assignment (FRANCIS)

```python
import pandas as pd
data2 = pd.read_csv('Cricket_data.csv')
data2.set_index('Unnamed: 0')
data2.loc[3, 'Player'] = 'David Warner'
```

```python
data2.loc[3,['Player','Salary']]
```

```python
gdata = data2.copy().groupby('Team')
```

```python
#1. To get the average salary of each team
gdata.mean().drop([ 'Unnamed: 0','Runs'], axis = 1).round(decimals = 2)
```

```python
#2. to get the palyer with the highest salary per team
hdata = gdata.max().drop(['Unnamed: 0','Runs'], axis = 1)
hdata.loc['Australia', 'Player'] = 'David Warner'
hdata
```

```python
#3. to get the players with the lowest runs per team
gdata.min().drop(['Unnamed: 0','Salary'] ,axis = 1)
```
