# MERGING, JOINING AND CONCATENATION

There are three major ways of joining dataframes. they are listed above 

```python
import pandas as pd
```

```python
left = pd.DataFrame({'id': [1,2,3,4,5], 'Name' : ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 'subject id':['sub1', 'sub2', 'sub4', 'sub6', 'sub5']})
right = pd.DataFrame({'id': [1,2,3,4,5], 'Name' : ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
                     'subject id':['sub2', 'sub4', 'sub3', 'sub6', 'sub5']})
print(left)
print()
print(right)
```

### concaternation
concatenation basically glues together dataframes. keep in mind that dimensions should match along the axis you are concatenating on. you can use **pd.concat** and pass in a list of dataframes to concatenate together

```python
center = pd.concat([left, right])
center
```

```python
#by default, axis = 0 and it concatenates on the rows, to concatenate on the columns change axis to 1
```

```python
center1 = pd.concat([left, right], axis = 1)
center1
```

### Merging Dataframes

```python
pd.merge(left, right, on = 'id')
```

```python
pd.merge(df1, df2, on = 'id no.')
```

```python
pd.merge(left, right, on = ['id', 'subject id'])
```

```python
pd.merge(left, right, on = ['id', 'subject id'], how = 'outer')
```

```python
pd.merge(left, right, on = ['id', 'subject id'], how = 'left')
```

```python
pd.merge(left, right, on = ['id', 'subject id'], how = 'right')
```

```python
pd.merge(df1, df2, on = ['id no.'], how = 'left')
```

### Joining 
Joining is a convenient method for combining the column of two potential differently indexed Dataframes into a single result Dataframe.

```python
df3 = pd.read_csv('Data11.csv')
df4 = pd.read_csv('Data12.csv')
df4
```

```python
df3.columns = ['Key', 'Name', 'Age']
df3
```

```python
df4.columns = ['Key', 'Address', 'Qualification']
df4.index = df4['Key']
df4_1 = df4.drop('Key', axis = 1)
df4_1
```

```python
df3.index = df3['Key']
df3_1 = df3.drop('Key', axis = 1)
df3_1
```

```python
df3_1.join(df4_1)
```

```python
df3_1.join(df4_1, how = 'right')
```
