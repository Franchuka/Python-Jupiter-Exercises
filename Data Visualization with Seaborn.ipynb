## Types of Plots

### 1. distplot()

```python
##Notes: One of the prefered plots for numeric data is histogram
#to create a histogram in seaborn, we use the distplot() function.
#histograms can only be used on univariate data and cannot show the relationships between two variable
```

```python
sns.distplot(tips['total_bill'])
```

```python
#note the above will no longer be used in future versions
sns.histplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

```python
tips['total_bill'].max()-tips['total_bill'].min()
```

```python
sns.histplot(titanic['age'])
```

### 2. jointplot

jointplot allows you to baically match up two distplot for bivariate data. with your choice of what **kind** of parameter to compare:
** "scatter"
** "reg"
** "hex"

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x = tips['total_bill'], y = tips['tip'])
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'age', y = 'fare', data = titanic)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

### 3. pairplot()

```python
sns1 = sns.pairplot(data = tips)
```

```python
sns.pairplot(data = tips, hue = 'sex')
```

### 4. Barplot()
this used to fnd the relationship between categorical data/column and numeric data/column . 
for each unique value in a categorical column, a bar is plotted, which by default, displays the mean value for the data in a numeric column specific by the bar plot. standard, the categorical data is on the x axis, while the numeric data is on the y axis

```python
sns.barplot(x = 'sex', y = 'tip', data = tips)
```

```python
rank = pd.read_csv('qs-world-university-rankings-2017-to-2022-V2.csv', usecols =['university', 'year', 'rank_display',
                                                                                 'score', 'country', 'city', 'region',
                                                                                 'type','international_students','size',
                                                                                 'faculty_count'])
```

```python
rank.info()
```

```python
rank.tail()
```

```python
sns.barplot(x= rank['region'], y = rank['score'])
```

```python
#by default the estimator for barplot is mean but you can change the estimator by using te argument
sns.barplot(x= rank['region'], y = rank['score'], estimator = np.std)
```

### countplot()
this is essentially the same as barplot except the estimator is explicitly counting the numbers of occurence. which is why we only pass the x value:

```python
sns.countplot(x='region', data= rank)
```
