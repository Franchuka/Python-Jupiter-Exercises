### Logical Operators in Pandas 
Pandas dataframe allow nfor boolean indexing, which is quite an efficient way to filter a dataframe for multiple conditions. In boolean indexing, boolean vectors generated based onn the conditions are used to filter the data. Multiple conditions involving the operators|(for or operation), & (for and operations), and ~(for not operation) can be grouped using parenthesis ().

```python
is_huge = df['Population']> 200 and df['Area'] > 5)
is_huge
# this will return an error
```

```python
# instead it is adviced to use either 
is_huge = (df['Population']> 200) & (df['Area'] > 5 )
df[is_huge]
```
